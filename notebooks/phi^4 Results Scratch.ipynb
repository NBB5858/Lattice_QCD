low epsilon (0.01) gives a consistently high acceptance rate. However, it makes the burn in extremely slow. 

moderate epsilon (0.05) gives an acceptance rate which is sometimes pitiful. However, when it is good, the burn in is generally good.

high epsilon (0.1) almost always gives a pitiful acceptance rate, but when it is good, the burn in is quite good.
